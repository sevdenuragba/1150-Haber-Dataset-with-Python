# Data Preprocessing

```python
import os
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem.isri import ISRIStemmer


nltk.download('stopwords')

def preprocess_text(text):
    # Tokenization
    sentences = sent_tokenize(text)
    tokens = [word_tokenize(sentence, language='turkish') for sentence in sentences]
    
    # Flatten the list of tokens
    flat_tokens = [token for sublist in tokens for token in sublist]
    
    # Removing stop words
    stop_words = set(stopwords.words('turkish'))
    filtered_tokens = [word for word in flat_tokens if word.isalnum() and word not in stop_words]
    
    print(stop_words)
    
    # Stemming
    isri_stemmer = ISRIStemmer()
    stemmed_tokens = [isri_stemmer.stem(word) for word in filtered_tokens]
    
    
    
    
    return stemmed_tokens

def preprocess_and_save(input_folder, output_folder):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Process each text file in the input folder
    for filename in os.listdir(input_folder):
        with open(os.path.join(input_folder, filename), 'r', encoding='cp1254') as file:
            raw_text = file.read()
            preprocessed_text = preprocess_text(raw_text)

            # Save preprocessed text to the output folder
            output_filepath = os.path.join(output_folder, f"preprocessed_{filename}")
            with open(output_filepath, 'w', encoding='utf-8') as output_file:
                output_file.write(' '.join(preprocessed_text))

# Example usage
input_folder = 'C:/Users/s_agb/Desktop/1150_input'
output_folder = 'C:/Users/s_agb/Desktop/1150_output'
preprocess_and_save(input_folder, output_folder)
```
